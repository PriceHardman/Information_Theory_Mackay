## 3.1: A First Inference Problem

Consider the following classical phyisics problem:

*Unstable particles are emitted from a source and decay at a distance $x$, a real number that has an exponential probability distribution with characteristic length $\lambda$. Decay events can be observed only if they occur in a windows extending from $x=1$cm to $x=20$cm. $N$ decays are observed at locations $\left\{x_{1},\cdots,x_{N}\right\}$. What is $\lambda$?*

Elementary classical statistics offers a few seemingly useful approaches: construct 'estimators' of the unknown parameters, 'fitting' the model to the data, etc.<br/>
Since the mean of an unconstrained exponential distribution is $\lambda$, it seems reasonable to examine the sample mean $\bar{x} = \sum_{n}{\frac{x_{n}}{N}}$ and see if an estimator $\hat{\lambda}$ could be obtained from it. While estimators can be constructed that are appropriate for sections of the range $1\leq \lambda \leq 20$, there is no obvious estimator that would work for the entire range.</br>
It is equally unsatisfactory to fit the density $P\left(x|\lambda\right)$ to a histogram derived from the data.<br/>
**What is the general solution to this problem and others like it? Is it always necessary to worry about finding the 'right' estimator (whatever that means)?**<br/><br/>
Before setting out in the direction of a solution, let us consider what we know: Given the mean $\lambda$, the probability density for a given point (unconstrained by the observation window $x \in \left[1,20\right]$) is $P\left(x|\lambda\right) =\frac{1}{\lambda}e^{\frac{-x}{\lambda}}$, $x \geq 0$. To take the observation window into account, we can constrain our probability density to $\left[1,20\right]$ by dividing our density by the constant $$Z\left(\lambda\right) = \int_{1}^{20}{\frac{1}{\lambda}e^{\frac{-x}{\lambda}}dx} = \left(e^{\frac{-1}{\lambda}}-e^{\frac{-20}{\lambda}}\right).$$
Thus,
$$P\left(x|\lambda\right) = \frac{1}{\lambda Z\left(\lambda\right)} e^{\frac{-x}{\lambda}}
= \frac{\frac{1}{\lambda}e^{\frac{-x}{\lambda}}}{e^{\frac{-1}{\lambda}}-e^{\frac{-20}{\lambda}}}$$ for $1 < x < 20$, and $0$ otherwise. Using this expression, we can plot the density $P\left(x|\lambda\right)$ as a function of $\lambda$ for different $x$ values, and even plot the bivariate density $P\left(x|\lambda\right)$ as a function of both $x$ and $\lambda$.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
%matplotlib inline

e = np.exp(1)

def Z(l):
    return (e**(-1/l)-e**(-20/l))

def P_x_given_l(x,l):
    return (1/(l*Z(l)))*(e**(-x/l))

def f(x,y):
    return x+y

lambdas = np.linspace(0,100,1000,endpoint=False)
L = np.linspace(1,100,100,endpoint=False)
X = np.linspace(1,20,100,endpoint=False)
X,L = np.meshgrid(X,L)
P = P_x_given_l(X,L)

f,ax = plt.subplots(nrows=1,ncols=2)

for x in [3,5,10,15]:
    ax[0].semilogx(lambdas,P_x_given_l(x,lambdas),label="x=%d" % x)
ax[0].set_xlabel(r'$\lambda$',fontsize=20)
ax[0].set_ylabel(r'$P\left(x|\lambda\right)$',fontsize=20)
ax[0].legend(loc=1, borderaxespad=0.)

ax[1] = plt.subplot(122,projection='3d')
ax[1].plot_surface(X,L,P)
ax[1].set_xlabel(r'$x$',fontsize=20)
ax[1].set_ylabel(r'$\lambda$',fontsize=20)
ax[1].xaxis._axinfo['label']['space_factor'] = 3
ax[1].yaxis._axinfo['label']['space_factor'] = 3

# NOTE: Mplot3d currently doesn't support log-scaled axes. :(

plt.show()
```

So far so good. However, now consider Bayes' Theorem:
\begin{align}
P\left(\lambda | \left\{x_{1},\cdots,x_{N}\right\}\right) &=
\frac{P\left(\left\{x\right\}|\lambda\right)P\left(\lambda\right)}{P\left(\left\{x\right\}\right)} \\
&\propto P\left(\left\{x\right\}|\lambda\right)P\left(\lambda\right) \\
&= \left(\prod_{n=1}^{N}{P\left(x_{n}|\lambda\right)}\right)P\left(\lambda\right) \\
&= \left(\prod_{n=1}^{N}{\left(\lambda Z\left(\lambda\right)\right)^{-1} e^{\frac{-x_{n}}{\lambda}}}\right)P\left(\lambda\right) \\
&= \left(\lambda Z\left(\lambda\right)\right)^{-N} \exp\left(-\sum_{1}^{N}{\frac{x_{n}}{\lambda}}\right)P\left(\lambda\right)
\end{align}<br/><br/>
Here, we are defining the probability of a hypothesis, given the data. Our calculations have resulted in an expression for this probability in terms of the datapoints $\left\{x_{1},\cdots,x_{N}\right\}$ and our hypothesis $\lambda$. The probability in this case represents a degree of belief. It should be noted that the hypothesis $\lambda$ that correctly describes the situation is **not** a stochastic variable, and the fact that the Bayesian uses a probability distribution $P$ does **not** mean that he thinks of the world as stochastically changing its nature between the states described by the different hypotheses. This notation is instead used to represent the Bayesian's **beliefs** about the mutually exclusive hypotheses (here, values of $\lambda$), of which only one is actually true.<br/>
The posterior distribution above represents the unique and complete solution to this problem. There is no need to invent 'estimators', nor do we need to invent criteria for comparing alternative estimators.

### Assumptions in Inference

Our inference is conditional upon our assumptions. For example, in the previous problem our result depends upon us assuming values for the prior probability $P\left(\lambda\right)$. Critics view such priors as a difficulty because they are 'subjective'. But how can it be otherwise? How can one perform inference without making assumptions? At the very least, Bayesian methods force one to make otherwise-tacit assumptions explicit.<br/><br/>
Once assumptions are made, inferences are objective and unique, able to be reproduced using the same information and assumptions. For example, given the assumptions listed above, $\mathcal{H}$, and the data $D$, everyone will agree about the posterior probability of the decay length $\lambda$:
$$P\left(\lambda|D,\mathcal{H}\right) = 
\frac{P\left(D|\lambda,\mathcal{H}\right)P\left(\lambda|\mathcal{H}\right)}
{P\left(D|\mathcal{H}\right)}$$<br/>
Secondly, when assumptions are explicit, they are both easier to criticize and easier to modify. Indeed, we can quantify the sensitivity of our inferences to the details of the assumptions. For instance, note in the plot above of $\lambda$ vs $P\left(x|\lambda\right)$, for $x=3$ there is a prominent peak in the distribution, whereas for higher values of $x$ such as $15$ there is no peak at all. In these higher $x$ cases, only low values of $x$ are ruled out, and no information is provided as to the relative probabilities of large values of $\lambda$. In these cases, our choice of the prior $P\left(\lambda\right)$ is more important in that it will affect the posterior probability to a greater degree.<br/><br/>
Third, when we are not sure which of various alternative assumptions is the most appropriate for the problem, we can treat this question as another inference task. Thus, given data $D$, we can compare alternative assumptions using Bayes' Theorem:
$$P\left(\mathcal{H}|D,I\right) = 
\frac{P\left(D|\mathcal{H},I\right)P\left(\mathcal{H}|I\right)}{P\left(D|I\right)},$$
where $I$ denotes te highest assumptions, which we are not questioning.<br/><br/>
Fourth, we can take into acount our uncertainty regarding such assumptions when we make subsequent preditions. Rather than choosing one particular assumption $\mathcal{H}^{*}$ and working out our predictions about some quantity $\mathbf{t}$, $P\left(\mathbf{t}|D,\mathcal{H}^{*},I\right)$, we obtain predictions that take into account our uncertainty about $\mathcal{H}$ by using the sume rule:
$$P\left(\mathbf{t}|D,I\right) = 
\sum_{\mathcal{H}}{P\left(\mathbf{t}|D,\mathcal{H},I\right)P\left(\mathcal{H}|D,I\right)}.$$<br/></br>
This is in contrast again with orthodox statistics, in which it is conventional to 'test' a default model, and then if the test 'accepts the model' at some 'significance level', to use exclusively that model to make predictions.

## 3.2: The Bent Coin

A bent coin with bias $p_{a}$ is tossed $F$ times. We observe a sequence $s$ of heads and tails (denoted by $a$ and $b$, respectively). We wish to know the bias of the coin, and predict the probability that the next toss will result in a head. We'll assume a uniform prior distribution and obtain a posterior distribution by multiplying by the likelihood. A critic might object 'Where did this prior come from?' We make no claim that this prior is in any way fundamental; it is simply a subjective assumption that every possible value of $p_{a}$ is equally likely. We will give the name $\mathcal{H}_{1}$ to our assumptions. Given $p_{a}$, the probability that $F$ tosses result in a sequence $\mathbf{s}$ containing $\left\{F_{a},F_{b}\right\}$ counts of heads and tails is
$$P\left(\mathbf{s}|p_{a},F,\mathcal{H}_{1}\right) = 
p_{a}^{F_{a}}\left(1-p_{a}\right)^{F_{b}}$$.<br/>
This model is assuming a uniform prior for $p_{a}$, whereby $P\left(p_{a}|\mathcal{H}_{1}\right) = 1$, with $p_{a} \in \left[0,1\right]$.

### Inferring Unknown Paramters

Given a string of length $F$, of which $F_{a}$ are heads and $F_{b}$ are tails, we are interested in (a) inferring what $p_{a}$ might be; (b) predicting whether the next character is an $a$ and $b$ (remember that predictions are expressed as probabilities, so 'predicting' is the same as computing the probability of an outcome).<br/><br/>
Assuming $\mathcal{H}_{1}$ to be true, using Bayes' Theorem and marginalizing over $p_{a}$ to calculate the nornmalization constant, we find by the Beta integral that
\begin{align}
P\left(p_{a}|s,F,\mathcal{H}_{1}\right) & =
\frac{P\left(s|p_{a},F,\mathcal{H}_{1}\right)P\left(p_{a}|\mathcal{H}_{1}\right)}{P\left(s|F,\mathcal{H}_{1}\right)} \\
&= \frac{\left(p_{a}^{F_{a}}\left(1-p_{a}\right)^{F_{b}}\right)\left(1\right)}{\int_{0}^{1}{p_{a}^{F_{a}}\left(1-p_{a}\right)^{F_{b}}dp_{a}}} \\
&= \left(p_{a}^{F_{a}}\left(1-p_{a}\right)^{F_{b}}\right)\left(\frac{F_{a}!F_{b}!}{\left(F_{a}+F_{b}+1\right)!}\right)^{-1} \\
&= \left(p_{a}^{F_{a}}\left(1-p_{a}\right)^{F_{b}}\right)\frac{\left(F_{a}+F_{b}+1\right)!}{F_{a}!F_{b}!}
\end{align}

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline

def flip(pa,n):
    """
    Flip a coin with bias pa n times, and return dict {Fa:_,Fb:_}
    with number of heads and tails that resulted.
    """
    Fa = scipy.stats.binom.rvs(p=pa,n=n)
    Fb = n - Fa
    return Fa,Fb

def P_pa_given_s(pa,Fa,Fb):
    """
    Given hypothetical bias pa and data {Fa,Fb},
    return posterior probability of the bias pa
    given the data.
    """
    P_s_given_pa_F = pa**Fa * (1.0-pa)**Fb
    P_s_given_F = np.math.factorial(Fa+Fb+1) / (np.math.factorial(Fa)*np.math.factorial(Fb))
    return P_s_given_pa_F / P_s_given_F

nrows = 3
ncols = 4
f,ax = plt.subplots(nrows=nrows,ncols=ncols,figsize=(10,10))

PA = [0.25,0.5,0.75]
for row in range(nrows):
    pa = PA[row]
    N = np.linspace(10,100,ncols)
    for col in range(ncols):
        n = N[col]
        Fa,Fb = flip(pa,n)
        X = np.linspace(0,1,1000,endpoint=False)
        Y = P_pa_given_s(X,Fa,Fb)
        ax[row][col].plot(X,Y,'r-')
        ax[row][col].axes.get_yaxis().set_ticks([])
        if col == 0:
            ax[row][col].set_ylabel(r"$p_{a}=$ %.2f" % pa,fontsize=20)
        if row == 0:
            ax[row][col].set_title(r"$F=$ %d" % n)
plt.suptitle(r'Distribution of $P\left(p_{a}|s,F,\mathcal{H}_{1}\right)$',fontsize=20)

plt.show()
```

Here we've plotted the posterior probability $P\left(p_{a}|s,F,\mathcal{H}_{1}\right)$ for three different values of $p_{a}$ and four different values of $F$. In each case, $F$ tosses of a coin with bias $p_{a}$ was simulated, and the results $\left\{F_{a},F_{b}\right\}$ were used in the calculation of $P\left(p_{a}|s,F,\mathcal{H}_{1}\right)$. Note that as $F$ increases, the posterior probability distribution narrows in on the bias $p_{a}$.

### From Inference to Prediction

With our expression for the probability of a heads $p_{a}$ given our previous string of tosses $\mathbf{s}$, our total number of tosses $F$, and our initial assumptions $\mathcal{H}_{1}$, the probability that the next toss is an $a$ (heads) is obtained by integrating over $p_{a}$:
$$P\left(a|s,F\right) = \int{P\left(a|p_{a}\right)P\left(p_{a}|s,F\right)dp_{a}}.$$
This has the effect of taking into account our uncertainty about $p_{a}$ when making our prediction. Noting that the probability of heads given $p_{a}$ is simply $p_{a}$, we evaluate the integral and find that
\begin{align}
P\left(a|s,F\right) &= \int{P\left(a|p_{a}\right)P\left(p_{a}|s,F\right)dp_{a}}\\
&= \int{\left(p_{a}\right)\left(p_{a}^{F_{a}}\left(1-p_{a}\right)^{F_{b}}\right)\frac{\left(F_{a}+F_{b}+1\right)!}{F_{a}!F_{b}!}dp_{a}}\\
&= \frac{\left(F_{a}+F_{b}+1\right)!}{F_{a}!F_{b}!}\int{\left(p_{a}^{F_{a}+1}\left(1-p_{a}\right)^{F_{b}}\right)dp_{a}}\\
&= \frac{\left(F_{a}+F_{b}+1\right)!}{F_{a}!F_{b}!}\cdot\frac{\left(F_{a}+1\right)!F_{b}!}{\left(F_{a}+F_{b}+2\right)!} \\
&= \frac{F_{a}+1}{F_{a}+F_{b}+2}.
\end{align}<br/>
This result is known as **Laplace's Rule**.

## 3.3: The Bent Coin and Model Comparison

Imagine that another theory for our data is introduced. Instead, it is supposed that the source of the data is not really a bent coin with a perfectly formed die with one face painted heads ('$a$'), and the other five painted tails ('$b$'). Thus, the parameter $p_{a}$, which in the original model $\mathcal{H}_{1}$ could take any value between $0$ and $1$, is according to the new hypothesis $\mathcal{H}_{0}$ not a free paramter at all, but rather is equal to $\frac{1}{6}$. **How can we compare these two models in light of the data?** We wish to infer how probable $\mathcal{H}_{1}$ is relative to $\mathcal{H}_{0}$.

### Model Comparison as Inference

In order to perform model comparison, we can use Bayes' Theorem to express how probable each model is in light of the data:
$$P\left(\mathcal{H}_{1}|s,F\right) = 
\frac{P\left(s|F,\mathcal{H}_{1}\right)P\left(\mathcal{H}_{1}\right)}{P\left(s|F\right)}$$<br/>
$$P\left(\mathcal{H}_{0}|s,F\right) = 
\frac{P\left(s|F,\mathcal{H}_{0}\right)P\left(\mathcal{H}_{0}\right)}{P\left(s|F\right)}$$<br/><br/>
The normalizing constant in both cases is the total probability of getting the observed data. If $\mathcal{H}_{1}$ and $\mathcal{H}_{0}$ are the only two models under consideration, this probability is given by the sum rule:
$$P\left(s|F\right) = 
P\left(s|F,\mathcal{H}_{1}\right)P\left(\mathcal{H}_{1}\right) +
P\left(s|F,\mathcal{H}_{0}\right)P\left(\mathcal{H}_{0}\right).$$<br/>
Thus, we have it that
$$P\left(\mathcal{H}_{1}|s,F\right) = 
\frac{P\left(s|F,\mathcal{H}_{1}\right)P\left(\mathcal{H}_{1}\right)}{P\left(s|F,\mathcal{H}_{1}\right)P\left(\mathcal{H}_{1}\right) +
P\left(s|F,\mathcal{H}_{0}\right)P\left(\mathcal{H}_{0}\right)}$$<br/>
$$P\left(\mathcal{H}_{0}|s,F\right) = 
\frac{P\left(s|F,\mathcal{H}_{0}\right)P\left(\mathcal{H}_{0}\right)}{P\left(s|F,\mathcal{H}_{1}\right)P\left(\mathcal{H}_{1}\right) +
P\left(s|F,\mathcal{H}_{0}\right)P\left(\mathcal{H}_{0}\right)}$$<br/><br/><br/>
To evaluate these hypotheses we need to assign values to the prior probabilities $P\left(\mathcal{H}_{1}\right)$ and $P\left(\mathcal{H}_{0}\right)$. To be safe, it seems reasonable in this case to assign them equal probabilities of $\frac{1}{2}$.<br/>
The quantity $P\left(s|F,\mathcal{H}_{1}\right)$ is a measure of how much the data favors model $\mathcal{H}_{1}$, and is called the *evidence* for model $\mathcal{H}_{1}$. We already encountered this quantity as the normalizing constant of the first inference we made: the inference of $p_{a}$ given the data. This is an instance of the following rule:<br/> 
**The evidence for a model is usually the normalizing constant of an earlier Bayesian inference.**<br/><br/>
We already calcuated the evidence for $\mathcal{H}_{1}$ as $$P\left(s|F,\mathcal{H}_{1}\right) = \int_{0}^{1}{p_{a}^{F_{a}}\left(1-p_{a}\right)^{F_{b}}dp_{a}} = \frac{F_{a}!F_{b}!}{\left(F_{a}+F_{b}+1\right)!}.$$<br/>
The evidence for model $\mathcal{H}_{0}$ is very simple because the model has no parameters to infer:
$$P\left(s|F,\mathcal{H}_{0}\right) = \left(\frac{1}{6}\right)^{F_{a}}\left(1-\frac{1}{6}\right)^{F_{b}} = \left(\frac{1}{6}\right)^{F_{a}}\left(\frac{5}{6}\right)^{F_{b}}$$<br/><br/>
Thus, the posterior probability ratio for model $\mathcal{H}_{1}$ and $\mathcal{H}_{0}$ is
\begin{align}
\frac{P\left(\mathcal{H}_{1}|s,F\right)}{P\left(\mathcal{H}_{0}|s,F\right)} &=
\frac{P\left(s|F,\mathcal{H}_{1}\right)P\left(\mathcal{H}_{1}\right)}{P\left(s|F\right)} \cdot \frac{P\left(s|F\right)}{P\left(s|F,\mathcal{H}_{0}\right)P\left(\mathcal{H}_{0}\right)} \\
&= \frac{P\left(s|F,\mathcal{H}_{1}\right)P\left(\mathcal{H}_{1}\right)}{P\left(s|F,\mathcal{H}_{0}\right)P\left(\mathcal{H}_{0}\right)} \\
&= \frac{P\left(s|F,\mathcal{H}_{1}\right)\cdot\frac{1}{2}}{P\left(s|F,\mathcal{H}_{1}\right)\cdot\frac{1}{2}} \\
&= \frac{P\left(s|F,\mathcal{H}_{1}\right)}{P\left(s|F,\mathcal{H}_{0}\right)} \\
&= \left(\frac{F_{a}!F_{b}!}{\left(F_{a}+F_{b}+1\right)!}\right)\cdot\left(\left(\frac{1}{6}\right)^{F_{a}}\left(\frac{5}{6}\right)^{F_{b}}\right)^{-1}
\end{align}<br/><br/>
It should be apparent that different $\left(F_{a},F_{b}\right)$ pairs favor different models, and that no outcome is completely incompatible with either model.<br/><br/>
Here we've plotted the results of nine simulated experiments. For three different values of $p_{a}$, three coin flipping experiments were run, in which a coin with bias $p_{a}$ was flipped $200$ times. The plots show the posterior probability ratio between models $\mathcal{H}_{1}$ and $\mathcal{H}_{0}$ as the result of the tosses accumulate.

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
%matplotlib inline

def flips(pa,size):
    """
    Simulates <size> flips of a coin with bias <pa>,
    returns an array of length <size>, where each element
    is a tuple containing the cumulative total of heads and tails thus far.
    """
    heads = scipy.stats.binom.rvs(p=pa,n=1,size=size) # binary array, 1=heads
    tails = np.bitwise_not(heads)+2 # invert so 1=tails
    totals = np.vstack((np.add.accumulate(heads),np.add.accumulate(tails))).T
    return [tuple(x) for x in totals]
    
def posterior_prob_ratio(Fa,Fb):
    """
    Given a headcount and tailcount (Fa,Fb), return the
    posterior probability ratio P(H1|s,F)/P(H0|s,F).
    """
    P_H1 = np.math.factorial(Fa)*np.math.factorial(Fb)/np.math.factorial(Fa+Fb+1)
    P_H0 = (1.0/6.0)**Fa * (5.0/6.0)**Fb
    return P_H1/P_H0

PA = [1/6,0.25,0.5] # the (hidden) biases for the experiements
nflips = 200
f,ax = plt.subplots(nrows=3,ncols=3,figsize=(12,10))
for col in range(3):
    pa = PA[col]
    for row in range(3):
        X = np.linspace(0,nflips,nflips)
        Y = np.log2([posterior_prob_ratio(*x) for x in flips(pa,nflips)])
        ax[row][col].plot(X,Y,'r-')
        ax[row][col].axhline(y=0,xmin=0,xmax=nflips)
        ax[row][col].set_ylim(-5.0,5.0)
        if row == 0:
            ax[row][col].set_title(r"$p_{a}=$ %.2f" % pa, fontsize=16)
        if col == 0:
            ax[row][col].set_ylabel(
                r"$\log_{2}\left(\frac{P\left(\mathcal{H}_{1}|s,F\right)}{P\left(\mathcal{H}_{0}|s,F\right)}\right)$",
                rotation='horizontal',
                fontsize=24
            )
            ax[row][col].yaxis.set_label_coords(-0.4,0.4)
plt.tight_layout()
plt.show()
```

Notice that in the case of $p_{a} = \frac{1}{6}$, the posterior ratio weighs in favor of $\mathcal{H}_{0}$, whereas in the cases of $p_{a} = 0.25$ the results only mildly favor $\mathcal{H}_{1}$. In the case of $p_{a} = 0.50$, however, the evidence very strongly favors $\mathcal{H}_{1}$.

## 3.4: An Example of Legal Evidence:

The following example illustrates that there is more to Bayesian inference than priors:<br/><br/>
**Two people have left traces of their own blood at a crime scene. A suspect, Oliver, is tested and found to have type 'O' blood. The blood types of the two traces at the scene were 'O' and 'AB'. Type 'O' blood has a frequency of $60%$ in the local population, whereas type 'AB' is rarer, with a frequency of $1%$. Do these data (that type 'O' and 'AB' blood were found at the scene) give evidence in favor of the proposition that Oliver was one of the two present at the crime?**<br/><br/>
To analyze this, let us:
* Denote the proposition $S$ as: '**The suspect Oliver and one unknown person were present.**'
* Denote the alternative proposition $\bar{S}$ as: '**Two unknown people from the population were present.**'
The prior here is the prior probability ratio between $S$ and $\bar{S}$. Our task here is simply to evaluate the contribution made by the data $D$. That is, to evaluate the likelihood ratio $\frac{P\left(D|S,\mathcal{H}\right)}{P\left(D|\bar{S},\mathcal{H}\right)}$. So in essence, we're evaluating here which proposition has a higher likelihood of resulting in the data.<br/><br/>
Since under proposition $S$ we already know that one of the traces will be of type 'O', then the probability of the data given $S$ is the probability that one unknown person drawn from the population has blood type 'AB':
$$P\left(D|S,\mathcal{H}\right) = p_{AB}$$<br/><br/>
The probability of the data given $\bar{S}$ is the probability that two unknown people drawn from the population have types 'O' and 'AB':
$$P\left(D|\bar{S},\mathcal{H}\right) = 2p_{O}p_{AB}$$<br/><br/>
In these equations, $\mathcal{H}$ denotes the assumptions that two people were present and left blood there, and the probability distribution of the blood groups of unknown people in an explanation is the same as the population frequencies.
Dividing, we obtain the likelihood ratio:
$$\frac{P\left(D|S,\mathcal{H}\right)}{P\left(D|\bar{S},\mathcal{H}\right)}
= \frac{p_{AB}}{2p_{O}p_{AB}} = \frac{1}{2p_{O}} = \frac{1}{2 \cdot 0.6} = 0.83$$
Thus, the data in fact provides weak evidence **against** the supposition that Oliver was present.
