# Chapter 21: Exact Inference by Complete Enumeration

We begin our discussion of methods for handling probabilities by discussing a brute-force inference method: complete enumeration of all hypotheses, and evaluation of their probabilities. This approach is an exact method, and the difficulty of carrying it out will motivate our smarter exact and approximate methods introduced in the following chapters.

## 21.1: The Burglar Alarm

Bayesian probability is sometimes called 'common sense, amplified'. When thinking about the following questions, please ask your common sense what it thinks the answers are; we will then see how Bayesian methods confirm our everyday intuition.

**Fred lives in LA and commutes 60 miles to work. While at work, he receives a phone call from his neighbor saying that Fred's burglar alarm is ringing. What is the probability that there was a burglar at Fred's house today? While driving home to investigate, Fred hears on the radio that there was a small earthquake that day near his home. 'Oh', he says, feeling relieved, 'it was probably the earthquake that set it off.' What is the probability that there was a burglar at Fred's house?**

Let's define some variables:
* Let $b$ be the event that a burglar was at Fred's house today.
* Let $a$ be the event that the alarm is ringing
* Let $p$ be the event that Fred receives a phone call from the neighbor reporting the alarm.
* Let $e$ be the event that a small earthquake took place near Fred's house today.
* Let $r$ be the event that a radio report of the earthquake is heard by Fred.
Then, the probability of all these events might factorize as follows:<br/>
$$P\left(b,e,a,p,r\right) = P\left(b\right)P\left(e\right)P\left(a|b,e\right)P\left(p|a\right)P\left(r|e\right),$$<br/>
since all events are plausibly independent, except for the following:<br/><br/>
* The alarm ringing could be dependent upon a burglar or an earthquake, so we say $P\left(a|b,e\right)$.
* Fred receiving a phone call could be dependent upon his alarm ringing, so we say $P\left(p|a\right)$.
* Hearing a report on the radio of an earthquake could be dependent upon an earthquake actually occurring, so we say that $P\left(r|e\right)$.<br/><br/>
Some plausible values for the probabilities are:
* **Burglar probability**:
$$P\left(b=1\right) = \beta, P\left(b=0\right) = 1-\beta,$$ (e.g. $\beta = 0.001$ gives a mean burglary rate of once every three years roughly).
* **Earthquake probability**: $$P\left(e=1\right)=\epsilon, P\left(e=0\right)=1-\epsilon,$$ with, e.g., $\epsilon=0.001$.
* **Alarm ringing probability**: We assume the alarm will ring if any of the following three events occurs: 
    * (a) A burglar enters the house and triggers the alarm. Let us assume alarm reliability of $a_{b}=0.99$, so that 99% of burglars trigger the alarm.
    * (b) An earthquake takes place and triggers the alarm. Let us assume that $a_{e}=0.01$, or 1% of alarms are triggered by earthquakes.
    * (c) Some other false alarm triggers the alarm. Let us assume a false alarm rate of $f=0.001$, so Fred has a false alarm from non-earthquakes about once every three years.
<br/><br/>
The probabilities of $a$ given $b$ and $e$ are then:
\begin{align} 
P\left(a=0|b=0,e=0\right) &= \left(1-f\right) &= 0.999, \\
P\left(a=0|b=1,e=0\right) &= \left(1-f\right)\left(1-a_{b}\right) &= 0.00999, \\
P\left(a=0|b=0,e=1\right) &= \left(1-f\right)\left(1-a_{e}\right) &= 0.98901, \\
P\left(a=0|b=1,e=1\right) &= \left(1-f\right)\left(1-a_{b}\right)\left(1-a_{e}\right) &= 0.0098901, \\
P\left(a=1|b=0,e=0\right) &= f &= 0.001, \\
P\left(a=1|b=1,e=0\right) &= 1-\left(1-f\right)\left(1-a_{b}\right) &= 0.99001, \\
P\left(a=1|b=0,e=1\right) &= 1-\left(1-f\right)\left(1-a_{1}\right) &= 0.01099, \\
P\left(a=1|b=1,e=1\right) &= 1-\left(1-f\right)\left(1-a_{b}\right)\left(1-a_{e}\right) &= 0.9901099
\end{align}<br/><br/>
We assume the neighbor would not call unless the alarm is ringing, and that the radio would not report an earthquake unless one had actually occurred.<br/><br/>
We can then answer the two questions about the burglar by computing the posterior probabilities of all hypotheses given the available information. Lets start by reminding ourselves that the probability that there is a burglar, before either $p$ or $r$ is observed, is $P\left(b=1\right) = \beta = 0.001$, and the probability that an earthquake took place is $P\left(e=1\right) = \epsilon = 0.001$, and that these two propositions are independent.<br/><br/>
First, when $p=1$, we know the alarm is ringing: $a=1$ (since the neighbor is an accurate reporter). The posterior probability of $b$ and $e$ becomes:
$$P\left(b,e|a=1\right) = \frac{P\left(a=1|b,e\right)P\left(b\right)P\left(e\right)}{P\left(a=1\right)}$$<br/>
The four possible values for the numerator are:
\begin{align}
P\left(a=1|b=0,e=0\right)P\left(b=0\right)P\left(e=0\right) &= 0.000998 \\
P\left(a=1|b=1,e=0\right)P\left(b=1\right)P\left(e=0\right) &= 0.000989\\
P\left(a=1|b=0,e=1\right)P\left(b=0\right)P\left(e=1\right) &= 0.000010979\\
P\left(a=1|b=1,e=1\right)P\left(b=1\right)P\left(e=1\right) &= 9.9\times 10^{-7}
\end{align}<br/><br/>
The normalizing constant is the sum of these four numbers, $P\left(a=1\right) = 0.002$, and the posterior probabilities are therefore:<br/><br/>
\begin{align}
P\left(b=0,e=0|a=1\right) &= 0.4993 \\
P\left(b=1,e=0|a=1\right) &= 0.4947 \\
P\left(b=0,e=1|a=1\right) &= 0.0055 \\
P\left(b=1,e=1|a=1\right) &= 0.0005
\end{align}<br/><br/>
This tells us the probabilities of various combinations of presences and absences of burglars and earthquakes, given that the alarm is ringing. To answer the question of how likely it was that a burglar was present, we marginalize over the earthquake variable $e$:
\begin{align}
P\left(b=0|a=1\right) = P\left(b=0,e=0|a=1\right)+P\left(b=0,e=1|a=1\right) &= 0.505 \\
P\left(b=1|a=1\right) = P\left(b=1,e=0|a=1\right)+P\left(b=1,e=1|a=1\right) &= 0.495 \\
\end{align}<br/><br/>
Thus, there is nearly a 50% chance that there was a burglar present. It is important to note that the variables $b$ and $e$, which were independent *a priori*, are now dependent. The posterior distribution is not a separable function of $b$ and $e$, and this fact is illustrated most simply by studying the effect of learning that $e=1$.<br/>
When we learn that $e=1$, the posterior distribution of $b$ is given by
$P\left(b|e=1,a=1\right) = P\left(b,e=1|a=1\right)/P\left(e=1,a=1\right)$, i.e. by dividing the bottom two rows by their sum $P\left(e=1|a=1\right)=0.0060$. The posterior probability of $b$ is then:
\begin{align}
P\left(b=0|e=1,a=1\right) &= 0.92 \\
P\left(b=1|e=1,a=1\right) &= 0.08 
\end{align}<br/><br/>
Where before there was nearly a 50-50 chance of a burglar being present, upon learning of the occurence of an earthquake, the probability of a burglar being present lowers to only 8%. This is in accordance with everyday intuition that the probability should reduce when Fred learns that an earthquake, an alternative explanation of the alarm, has happened.

### Explaining Away

This phenomenon, that one of the possible causes ($b=1$) of some data (the data in this case being that $a=1$) becomes **less** probable when another of the causes ($e=1$) becomes more probable, even though those two causes were independent variables *a priori*, is known as **explaining away**. Explaining away is an important feature of correct inferences, and one way that any artificial intelligence should replicate.<br/>
If we believe that the neighbor and the radio report are unreliable, the calculations become more complex, but the explaining-away effect persists. The arrival of the earthquake report $r$ simultaneously makes it *more* probable that the alarm is truly ringing, and *less* probable that the burglar was present.<br/>
In summary, we solved the inference questions about the burglar by enumerating all four hypotheses about the variables $\left(b,e\right)$, finding their posterior probabilities, and marginalizing to obtain required inferences about $b$.

## 21.2: Exact Inferences for Continuous Hypothesis Spaces

Many of the hypothesis spaces we will consider are naturally thought of as continuous. However, in any practical computer implementation, such continuous spaces will be necessarily discretized, and so can be enumerated on a grid of parameter values, for example.

### A Two-Parameter Model

Lets look at the Gaussian distribution as an example of a model with a two-dimensional hypothesis space. The one-dimensional Gaussian distribution is parameterized by a mean $\mu$ and a standard deviation $\sigma$:
$$P\left(x|\mu,\sigma\right) = 
\frac{1}{\sqrt{2\pi\sigma}}
\exp\left(-\frac{\left(x-\mu\right)^{2}}{2\sigma^{2}}\right) \equiv \mathcal{N}\left(x;\mu,\sigma^{2}\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def norm(x,mu,sigma):
    return (1/np.sqrt(2*np.pi*sigma**2))*np.exp(-((x-mu)**2)/(2*sigma**2))

x = np.linspace(-10,10,100)
n_sigma = 10
n_mu = 10
sigmas = np.linspace(1,10,num=n_mu,endpoint=False)
mus = np.linspace(-10,10,num=n_sigma,endpoint=False)

f,ax = plt.subplots(nrows=n_sigma,ncols=n_mu,figsize=(10,10))

for i in range(n_sigma):
    for j in range(n_mu):
        sigma, mu = sigmas[-(i+1)], mus[j]
        ax[i][j].plot(x,norm(x,mu,sigma))
        ax[i][j].axes.get_xaxis().set_visible(False)
        ax[i][j].axes.get_yaxis().set_visible(False)
        ax[i][j].set_ylim([0,0.3])
plt.show()
```

Above we see 100 hypotheses for the mean and standard deviation of a one-dimensional Gaussian. These hypotheses are evenly spaced in a 10x10 square grid.
